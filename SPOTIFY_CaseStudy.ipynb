### SPOTIFY CASE STUDY - EXPLORATORY DATA ANALYSIS

```python
#importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
df_tracks=pd.read_csv(r"G:\SpotifyCaseStudy\tracks.csv")
df_tracks.head()
```

#### 1. Checking null values

```python
df_tracks.isnull().sum()
```

#### 2. Getting information on the dataset and checking the datatypes

```python
df_tracks.info()
```

Since there are only 71 null values in the name column, we will remove those records, as it won't affect our dataset.

```python
df_tracks=df_tracks[~df_tracks["name"].isnull()]
```

```python
df_tracks.isnull().sum()
```

#### 3. 10 least populart songs/artists on Spotify

```python
least_popular=df_tracks.sort_values("popularity",ascending=True).head(10)
least_popular
```

#### 4. Getting descriptive statistics of the dataset

```python
df_tracks.describe()
```

```python
df_tracks.describe().transpose()
```

#### 5. Top 10 most popular songs

```python
most_popular=df_tracks[df_tracks['popularity']>90].sort_values('popularity',ascending=False).head(10)
most_popular
```

```python
#checking which artist is present at the 15th row
df_tracks[["artists"]].iloc[15]
```

#### 7. Converting the duration from miliseconds to seconds

```python
# changing the columns name first
df_tracks.rename(columns={'duration_ms':'duration_sec'},inplace=True)
```

```python
df_tracks.columns
```

```python
#applying formula to change milisec to second
df_tracks["duration_sec"]=df_tracks["duration_sec"].apply(lambda x: round(x/1000))
df_tracks.head()
```

#### 8. Dropping redundant columns

```python
df_tracks.drop(['key','explicit','mode'],axis=1,inplace=True)
```

#### 9. Plotting correlation matrix

```python
plt.figure(figsize=[15,10])
sns.heatmap(df_tracks.corr(),annot=True,cmap="Greens")
plt.title("Correlation Matrix",fontdict={'fontsize':15})
plt.show()
```

#### 10. Based on the matrix, let's plot Regression plots for some variables.

a. Loudness vs Energy

```python
#sampling the data. Taking 0.4% of the total dataset
len(df_tracks)*0.004
```

```python
df_sample=df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
len(df_sample)
```

```python
df_sample.head()
```

```python
plt.figure(figsize=[10,5])
sns.regplot(data=df_sample,y='loudness',x='energy',line_kws={'color':'yellow'})
plt.xlabel("Energy",fontdict={'fontsize':15})
plt.ylabel("Loudness",fontdict={'fontsize':15})
plt.title("Loundness vs Energy",fontdict={'fontsize':20})
plt.show()
```

b) Loudness vs acousticness

```python
plt.figure(figsize=[10,5])
sns.regplot(data=df_sample,y='popularity',x='acousticness',line_kws={'color':'yellow'})
plt.xlabel("Energy",fontdict={'fontsize':15})
plt.ylabel("Popularity",fontdict={'fontsize':15})
plt.title("Popularity vs Acousticness",fontdict={'fontsize':20})
plt.show()
```

c) Valence vs Danceability

```python
plt.figure(figsize=[10,5])
sns.regplot(data=df_sample,y='valence',x='danceability',line_kws={'color':'yellow'})
plt.xlabel("Danceability",fontdict={'fontsize':15})
plt.ylabel("Valence",fontdict={'fontsize':15})
plt.title("Valence vs Danceability",fontdict={'fontsize':20})
plt.show()
```

#### 11. Extracting year from the date column

```python
#creating new year column
df_tracks["year"]=pd.DatetimeIndex(df_tracks['release_date']).year
```

```python
df_tracks.head()
```

Plotting number of songs per year since 1922

```python
plt.figure(figsize=[15,8])
sns.countplot(data=df_tracks,x="year")
plt.yscale("log")
plt.xticks(rotation=90)
plt.show()
```

```python
#since there is a lot of data, we are unable to see any trend, let's plot a histogram instead
plt.figure(figsize=[15,7])
sns.histplot(data=df_tracks,x="year",bins='auto',discrete=True)
plt.title("Yearwise song count",fontdict={'fontsize':15})
plt.ylabel("Count",fontdict={'fontsize':15})
plt.xlabel("Year",fontdict={'fontsize':15})

plt.show()
```

```python
sns.distplot(df_tracks["year"])
```

#### 12. Analysing Year vs Duration

a) Bar Plot : Duration vs Year

```python
plt.figure(figsize=[15,8])
sns.barplot(data=df_tracks,x="year",y="duration_sec",errwidth=False)
plt.xticks(rotation=90)
plt.title("Duration vs Year",fontdict={'fontsize':13})
plt.ylabel("Duration in sec",fontdict={'fontsize':13})
plt.xlabel("Year",fontdict={'fontsize':13})
plt.show()
```

b) Line plot : Duration vs Year

```python
plt.figure(figsize=[15,8])
sns.lineplot(data=df_tracks,x="year",y="duration_sec",err_kws=None)
plt.xticks(rotation=90)
plt.title("Duration vs Year",fontdict={'fontsize':15})
plt.ylabel("Duration in sec",fontdict={'fontsize':13})
plt.xlabel("Year",fontdict={'fontsize':13})
plt.show()
```
